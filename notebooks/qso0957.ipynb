```python
from PIL import Image
import os
import numpy as np
from mpl_toolkits.axes_grid1.inset_locator import zoomed_inset_axes
from mpl_toolkits.axes_grid1.inset_locator import mark_inset
import matplotlib.pyplot as plt
image_pil = Image.open('../figures/qso0957.jpg')
```

```python
image = np.array(image_pil)
plt.figure(figsize=(8, 8))
plt.imshow(image_pil)
ax = plt.gca()
```

```python

y1, x1 = 1300, 1850
y2, x2 = 1700, 2250

plt.figure(figsize=(20, 20))
extent=[0, image.shape[1], 0, image.shape[0]]
plt.imshow(image, extent=extent)

ax = plt.gca()

axins = zoomed_inset_axes(ax, 4, loc=4) # zoom = 6
axins.imshow(image, interpolation="nearest", extent=extent)
plt.setp(list(axins.spines.values()), linewidth=3, linestyle="-", color="r");
axins.axes.xaxis.set_visible(False)
axins.axes.yaxis.set_visible(False)
# axins.axis("off")
axins.set_xlim(x1, x2)
axins.set_ylim(y1, y2)
mark_inset(ax, axins, loc1=1, loc2=3, fc="none", ec="r")
ax.axis("off")

axins.annotate("B", [2100, 1450], color="w", fontsize=20)
axins.annotate("A",  [2000, 1600], color="w", fontsize=20)
axins.annotate("G",  [2050, 1500], color="w", fontsize=20)
plt.savefig("../figures/zoomed_in_qso0957.jpg", facecolor="w", bbox_inches="tight")
```
